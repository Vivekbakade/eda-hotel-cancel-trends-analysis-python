# Hotel Booking Data Analysis

## Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

# Set professional plot style
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)
```

## Loading the Dataset

```python
df = pd.read_csv("hotel_bookings 2.csv")
```

1. Dealing with Missing Values

```python
df.isnull().sum()
```

```python
df.drop(['company', 'agent'], axis=1, inplace=True)
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df = df[df['adr'] < 5000]
```

## Data Analysis and Visualizations

How Many Booking Were Cancelled

```python
cancelled_perc = df['is_canceled'].value_counts(normalize=True)
print(cancelled_perc)

plt.figure(figsize=(8, 6))
plt.title('Reservation Status Count', fontsize=16)
plt.bar(['Not Cancelled', 'Cancelled'], df['is_canceled'].value_counts(), color=['#4C72B0', '#C44E52'], width=0.6)
plt.ylabel('Number of Reservations')
plt.show()
```

Reservation Status in Different Hotels

```python
plt.figure(figsize=(10, 6))
ax1 = sns.countplot(x='hotel', hue='is_canceled', data=df, palette='viridis')
plt.title('Reservation Status in Different Hotels', fontsize=16)
plt.xlabel('Hotel Type')
plt.ylabel('Number of Reservations')
plt.legend(['Not Cancelled', 'Cancelled'])
plt.show()
```

Average Daily Rate in City and Resort Hotel

```python
resort_hotel = df[df['hotel'] == 'Resort Hotel']
city_hotel = df[df['hotel'] == 'City Hotel']

resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()

plt.figure(figsize=(20, 8))
plt.title('Average Daily Rate in City and Resort Hotel', fontsize=20)
plt.plot(resort_hotel.index, resort_hotel['adr'], label='Resort Hotel', color='#4C72B0')
plt.plot(city_hotel.index, city_hotel['adr'], label='City Hotel', color='#DD8452')
plt.xlabel('Date')
plt.ylabel('Average Daily Rate (ADR)')
plt.legend(fontsize=12)
plt.show()
```

Reservation Status per Month

```python
df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize=(12, 7))
ax1 = sns.countplot(x='month', hue='is_canceled', data=df, palette='magma')
plt.title('Reservation Status per Month', fontsize=16)
plt.xlabel('Month')
plt.ylabel('Number of Reservations')
plt.legend(['Not Cancelled', 'Cancelled'])
plt.show()
```

ADR for Each Month

```python
plt.figure(figsize=(12, 7))
plt.title('ADR for Each Month', fontsize=16)
sns.barplot(x='month', y='adr', data=df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index(), palette='rocket')
plt.xlabel('Month')
plt.ylabel('Total ADR')
plt.show()
```

Top 10 Countries with Reservation Cancelled

```python
cancelled_data = df[df['is_canceled'] == 1]
top_10_countries = cancelled_data['country'].value_counts()[:10]

plt.figure(figsize=(10, 10))
plt.title('Top 10 Countries with Reservation Cancelled', fontsize=16)
plt.pie(top_10_countries, autopct='%.2f', labels=top_10_countries.index, colors=sns.color_palette('pastel'))
plt.show()
```

Average Daily Rate - Cancelled vs Not Cancelled

```python
cancelled_df_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace=True)
cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

not_cancelled_data = df[df['is_canceled'] == 0]
not_cancelled_df_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace=True)
not_cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

plt.figure(figsize=(20, 8))
plt.title('Average Daily Rate - Cancelled vs Not Cancelled', fontsize=20)
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label='Not Cancelled', color='#4C72B0')
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label='Cancelled', color='#C44E52')
plt.xlabel('Date')
plt.ylabel('Average Daily Rate (ADR)')
plt.legend(fontsize=12)
plt.show()
```
